```python
import numpy as np

def price_tarn(S0, K, rd, rf, sigma, T, target, N):
    dt = 1/12
    steps = int(T * 12)
    drift = (rd - rf - 0.5 * sigma**2) * dt
    vol   = sigma * np.sqrt(dt)

    payoffs = []

    for i in range(N):
        S = S0
        cumC = 0
        payoff = 0
        alive = True

        for t in range(1, steps + 1):
            if not alive:
                break

            Z = np.random.standard_normal()
            S = S * np.exp(drift + vol * Z)

            c = max(K - S, 0)

            if cumC + c >= target:
                c = target - cumC
                alive = False

            cumC += c
            disc = np.exp(-rd * t * dt)
            payoff += c * disc

        payoffs.append(payoff)

    return np.mean(payoffs)


if __name__ == "__main__":
    # Parameters
    S0 = 100
    K = 100
    rd = 0.02
    rf = 0.01
    sigma = 0.2
    T = 2
    target = 0.1
    N = 10000

    price = price_tarn(S0, K, rd, rf, sigma, T, target, N)
    print(f"Prix TARN : {price*100:.2f}%")
```
